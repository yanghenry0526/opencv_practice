```python
import cv2

global img
def onMouse(event , x , y , flags , param):
    if img.ndim != 3 : #不等於三就是灰階
        print((x,y), end='')
        print('Gray ' , img[x,y])
    else:
        print((x,y), end='')
        print('RGB ' , img[x,y,2] , img[x,y,1] , img[x,y,0])  #因為呈現方式是BGR所以倒轉print
        
    
filename = input('enter filename :')
img = cv2.imread(filename)

print('rows : ' , img.shape[0])
print('columns : ' , img.shape[1])

cv2.namedWindow(filename)
cv2.setMouseCallback(filename , onMouse)
cv2.imshow(filename , img)
cv2.waitKey(0)
```
